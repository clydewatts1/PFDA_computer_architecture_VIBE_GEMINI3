## How the data is downloaded (Problem 1)
- Uses `yfinance.download` with `interval="1h"` and a window of the previous five days.
- Combines all tickers into one DataFrame with a `Ticker` column, then saves a single CSV file.

```python
# Example: run the download from the notebook (optional)
import subprocess
subprocess.run([sys.executable, "faang.py", "download"], check=False)
```

## How the plot is created (Problem 2)
- Reads the latest CSV from `data/`.
- Normalizes per ticker by dividing each series by its first Close value (start=1.0).
- Plots all tickers on one chart with labels, legend, grid, and a date title.

```python
# Example: run the plot from the notebook (optional)
import subprocess
subprocess.run([sys.executable, "faang.py", "plot"], check=False)
```

## Script (Problem 3)
- Linux shebang: `#!/usr/bin/env python3`.
- Marked executable on Linux using `chmod +x faang.py`.
- CLI supports `download`, `plot`, and `all` (default: `all`).

## Workflow (Problem 4)
The workflow file `.github/workflows/faang.yml` includes:
- `on.schedule.cron: "0 6 * * 6"` — runs Saturdays at 06:00 UTC.
- `actions/checkout@v4` — fetches repository.
- `actions/setup-python@v5` — installs Python 3.11.
- `pip install -r requirements.txt` — installs dependencies.
- `chmod +x faang.py` and `./faang.py all` — runs the script.
- `actions/upload-artifact@v4` — uploads `data/*.csv` and `plots/*.png` artifacts.

## Notes
- Normalization choice: start each ticker at 1.0 to compare relative performance.
- Combined CSV makes plotting and processing simpler.
- GitHub Actions uses UTC; adjust cron if a different local time is needed.
